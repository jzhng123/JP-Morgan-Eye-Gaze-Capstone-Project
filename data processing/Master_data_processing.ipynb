```python
#data regoin generator by Lizhou He
def get_x(x):
    x = x * 1440
    if 165<= x < 165+(503-165)/2 :
        return 1
    elif 165+(503-165)/2 <= x < 503:
        return 2
    elif 503 <= x < 503 + (503-165)/2:
        return 3
    elif 503 + (503-165)/2 <=x < 842:
        return 4
    elif 1099 <= x  < 1266:
        return 5
    else:
        return 0
def get_y(x,y):
    y = 855 - 810*y
    if get_x(x) in [1,2,3,4]:
        if 710 - (710-325)/4<= y < 710 :
            return 1
        elif 710 - 2*(710-325)/4 <= y < 710 - (710-325)/4:
            return 2
        elif 710 - 3*(710-325)/4 <= y < 710 - 2*(710-325)/4:
            return 3
        elif  325 <=y < 710 - 3*(710-325)/4:
            return 4
        else:
            return 0
    elif get_x(x)  == 5:
        if 710 - (710-333)/4<= y < 710 :
            return 1
        elif 710 - 2*(710-333)/4 <= y < 710 - (710-333)/4:
            return 2
        elif 710 - 3*(710-333)/4 <= y < 710 - 2*(710-333)/4:
            return 3
        elif  333 <=y < 710 - 3*(710-333)/4:
            return 4
        else:
            return 0
    else:
        return 0
def get_region(x,y):
    if get_y(x,y) == 0:
        return (0,0)
    else:
        return (get_x(x),get_y(x,y))
     
```

```python
#data region assginment 
#eye tracking region 
def df_processing(df):
    df['region'] = None
    df['region'] = df.apply(lambda x: get_region(x['BPOGX'],x['BPOGY']),axis=1)
    #cursor click region
    df['c_region']= None
    df['c_region'] = df.apply(lambda x: get_region(x['CX'],x['CY']),axis=1)
    #helper function to return the x or y coordinate
    def returnx(x,i):
        return x[i]
    #eye tracking region x coordinate
    df['region_x'] = None
    df['region_x'] = df.apply(lambda x: returnx(x['region'],0),axis=1)
    #eye tracking region y coordinate
    df['region_y'] = None
    df['region_y'] = df.apply(lambda x: returnx(x['region'],1),axis=1)
    #cursor click region x coordinate to check whether a decision is made
    df['c_region_x'] = None
    df['c_region_x'] = df.apply(lambda x: returnx(x['c_region'],0),axis=1)
    #cursor click region y coordinate 
    df['c_region_y'] = None
    df['c_region_y'] = df.apply(lambda x: returnx(x['c_region'],1),axis=1)
    return(df)
```

```python
#main function to output the decision time in each cell
#algorithm: sum the time between the mouse clicks on the decision-making area of the screen - 1,2,3,4 on the right
def get_decision_time(df,starttime, endtime,correct,question, player_experience,player_id):
   
    #selecting data only in this question
    df = df[(df.iloc[:,3]>=starttime) & (df.iloc[:,3]<=endtime)]
    #selecting decision making data 
    df_decision=df[(df["CS"]==1) & (df["c_region_x"]==5)]
    #collect decision making time
    decision_time = df_decision.iloc[:,3]
    #count the number decision making
    num_decision = len(df_decision)
    #create the 3D array for decision - 1st dimension: decision_ID, 2nd & 3rd: coordinate
    tm_array=[]
    for i in range(num_decision):
        #check the correctness of the answer
        if i > len(correct)-1: # not a valid decision
            correctness = 0
        else:
            correctness = correct[i]
        output = np.zeros((6,6))
        x=np.zeros(4)
        y=np.zeros(4)
        #get previous decision time
        if i == 0: # special case for the first decision of any question
            decision_start=starttime
            decision_end=decision_time.iloc[i]
        else:
            decision_start=decision_time.iloc[i-1]
            decision_end=decision_time.iloc[i]
            output = get_cell_time(df,output,decision_start, decision_end)
        for k in range(1,5):
            x[k-1]=round(np.sum(output[k]),3)
            y[k-1]=round(np.sum(output[:,k]),3)
        #dict output all the features
        combined = {
            "player_ID":player_id,
            "player_experience":player_experience,
            "Question_ID": question,
            "correctness":correctness,
            "start":decision_start,
            "end":decision_end,
            "(0,0)":output[0,0],
            "(1,1)":output[1,1],
            "(1,2)":output[1,2],
            "(1,3)":output[1,3],
            "(1,4)":output[1,4],
            "(2,1)":output[2,1],
            "(2,2)":output[2,2],
            "(2,3)":output[2,3],
            "(2,4)":output[2,4],
            "(3,1)":output[3,1],
            "(3,2)":output[3,2],
            "(3,3)":output[3,3],
            "(3,4)":output[3,4],
            "(4,1)":output[4,1],
            "(4,2)":output[4,2],
            "(4,3)":output[4,3],
            "(4,4)":output[4,4],
            "(5,1)":output[5,1],
            "(5,2)":output[5,2],
            "(5,3)":output[5,3],
            "(5,4)":output[5,4],
            "x1": x[0],
            "x2": x[1],
            "x3": x[2],
            "x4": x[3],
            "y1": y[0],
            "y2": y[1],
            "y3": y[2],
            "y4": y[3],
            "total": round(np.sum(output),3)
        }
        if combined["total"] > 1:
            tm_array.append(combined) #3D array for decision - 1st dimension: decision_ID, 2nd & 3rd: coordinate
    return tm_array
def get_cell_time(df,output,start, end): #helper function to capture the time
    for i in range(len(output)):
        for j in range(len(output[i])):
            tf=df[(df.iloc[:,3]>=start) & (df.iloc[:,3]<=end) & (df["region"]==(i,j))]
            output[i,j] = round(0.01666*len(tf),3) # use the fact that the data is captured at 60 hertz
    return(output)
    
```
